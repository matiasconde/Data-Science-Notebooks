## Deep Neural Networks - Building a Handwritten Digits Classifier

### In this Project we'll: 
    
    
- Explore why image classification is a hard task
    
- Observe the limitations of traditional machine learning models for image classification
      
- Train, test, and improve a few different deep neural networks for image classification

dataset link: http://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits

```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error, make_scorer, accuracy_score

%matplotlib inline
```

```python
digits = load_digits()
data = pd.DataFrame(digits.data)
target = pd.Series(digits.target)
```

### Let's see some images:

```python
images = [0,100,200,300,1000,1100,1200,1300]
np_images = []

for i in images:
    image = data.iloc[i]

    values = image.values
    np_images.append(values.reshape(8,8))
    
```

```python
fig = plt.figure(figsize=(10,10))

for i,image in enumerate(np_images):
    ax = fig.add_subplot(2,4,i+1)
    ax.imshow(image, cmap='gray_r')
```

### Split up the data into train and test sets: 
- train --> 80%
- test --> 20%

```python
train_range = round(data.shape[0]*0.8)
test_range = round(data.shape[0]*0.2)
print(train_range, "rows for the train set")
print(test_range, "rows for the test set")
data.shape[0] == train_range + test_range
```

```python
def train_knn(k,data, train_percentage):
    """k: neighbors for k-nearest neighbors (knn)"""
    train_range = round(data.shape[0]*train_percentage)
    knn = KNeighborsClassifier(k)
    X_train = data.iloc[0:train_range]
    y_train = target.iloc[0:train_range]
    knn.fit(X_train, y_train)
    return knn
    
def test_knn(data, knn, train_percentage):
    train_range = round(data.shape[0]*train_percentage)
    X_test = data.iloc[train_range:]
    predictions = knn.predict(X_test)
    return predictions
    
def cross_validate_knn(folds,knn, data, target, train_percentage):
    kf = KFold(folds,shuffle=True,random_state=1)
    test_index = round(data.shape[0]*train_percentage)
    y_test = target.iloc[test_index:]
    mses = cross_val_score(knn,data,\
        target,scoring=make_scorer(mean_squared_error),cv=kf)
    return mses
```

```python
#try the model
knn = train_knn(4,data,0.8)
predictions = test_knn(data,knn,0.8)
mean_mses = cross_validate_knn(4,knn,data,target, 0.8).mean()
mean_mses
```

```python
x1_values = pd.Series(predictions)
x2_values = target.iloc[1438:].reset_index(drop=True)
comparaciones = pd.concat([x1_values,x2_values],axis = 1)
comparaciones.columns = ["predictions", "true_values"]
comparaciones["position"] = comparaciones.index
matches = (comparaciones["predictions"] == comparaciones["true_values"]).sum()
unmatches = comparaciones.shape[0] - matches
print(matches, "number of matches")
print(unmatches, "number of unmatches")
```

```python
# Coordinates of unmatches.
unmatches = comparaciones.loc[comparaciones["predictions"] != comparaciones["true_values"]]
unmatch_coord = []
for value in unmatches[["position", "predictions"]].values:
    x = value[0]
    y = value[1]
    unmatch_coord.append((x,y))
```

```python
from matplotlib.patches import Ellipse

elipses = []
for coord in unmatch_coord:
    ell = Ellipse(coord,15,0.5, color ='blue', fill=False)
    elipses.append(ell)
    
# Plotting the results:

fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(1,1,1)

plt.scatter(comparaciones.index, comparaciones["predictions"], label="Bad predictions",color="r")
plt.scatter(comparaciones.index, comparaciones["true_values"], label="Correct predictions", color="g")

for elipse in elipses:
    ax.add_patch(elipse)
    
plt.legend(loc= 'lower left')
#plt.legend(loc= 'lower right')

plt.show()
```

### Downsides to using k-nearest neighbors:

- High memory usage (for each new unseen observation, many comparisons need to be made to seen observation).
- No model representation to debug and explore.

## Let's try neural networks:

```python
def train_mlp(k_l_tuple,data, train_percentage, activation_function = 'relu'):
    """
    - Predictor: Multi-layer Perceptron classifier (mlp)
    - k_l_tuple: tuple with (n1,n2,...) where each n1 is 
    the number of neurones, and each position is the layer
    with ni neurons, i:1..n
    - Activation functions: identity, logistic, tanh, relu. 
    """
    train_range = round(data.shape[0]*train_percentage)
    mlp = MLPClassifier(hidden_layer_sizes = k_l_tuple, \
                        max_iter = 1000, activation = activation_function)
    X_train = data.iloc[0:train_range]
    y_train = target.iloc[0:train_range]
    mlp.fit(X_train, y_train)
    return mlp
    
def test_mlp(data, mlp, train_percentage):
    train_range = round(data.shape[0]*train_percentage)
    X_test = data.iloc[train_range:]
    predictions = mlp.predict(X_test)
    return predictions
    
def cross_validate_mlp(folds, mlp, data, target, train_percentage):
    kf = KFold(folds,shuffle=True,random_state=1)
    test_index = round(data.shape[0]*train_percentage)
    y_test = target.iloc[test_index:]
    mses = cross_val_score(mlp,data,\
        target,scoring=make_scorer(accuracy_score),cv=kf)
    return mses
```

### 1. Neural network with different number of neurons and a single hidden layer:

```python
#try the model with different numbers of neurons and 4-fold cross val:

configuration = [(8,),(16,),(32,),(64,),(128,),(256,)]
mses = []
for i in configuration:
    mlp = train_mlp(i,data,0.8,'relu')
    predictions = test_mlp(data,mlp,0.8)
    scores = cross_validate_mlp(4,mlp,data,target, 0.8)
    mean_mses = scores.mean()
    mses.append(mean_mses)
    print(mean_mses, "mean of mean square error in cross val")
```

```python
models = pd.Series(configuration)
mses = pd.Series(mses)
ranking_models = pd.concat([models,mses], axis=1)
ranking_models.columns = ["Model", "Ranking"]
ranking_models.sort_values("Ranking", ascending=False,\
                           inplace=True)
ranking_models.reset_index(drop=True)
```

### 2. Neural network with 64 neurons and two hidden layers:

```python
#try the model with 64 neurons, two layers and 4-fold cross val:

configuration = [(64,64)]
mses = []
for i in configuration:
    mlp = train_mlp(i,data,0.8,'relu')
    predictions = test_mlp(data,mlp,0.8)
    scores = cross_validate_mlp(4,mlp,data,target, 0.8)
    mean_mses = scores.mean()
    mses.append(mean_mses)
    print(mean_mses, "mean of mean square error in cross val")
```

### 3. Neural network with 10, 64 and 128 neurons, three hidden layers and 6-fold cross validation:

```python
#try the model with different numbers of neurons, three hidden layers
#and 6-fold cross val:

configuration = [(10,10,10),(64,64,64),(128,128,128)]

mses = []
for i in configuration:
    mlp = train_mlp(i,data,0.8,'relu')
    predictions = test_mlp(data,mlp,0.8)
    scores = cross_validate_mlp(6,mlp,data,target, 0.8)
    mean_mses = scores.mean()
    mses.append(mean_mses)
    print(mean_mses, "mean of mean square error in cross validation")
```

```python
models = pd.Series(configuration)
mses = pd.Series(mses)
ranking_models = pd.concat([models,mses], axis=1)
ranking_models.columns = ["Model", "Ranking"]
ranking_models.sort_values("Ranking", ascending=False, inplace=True)
ranking_models.reset_index(drop=True)
```
